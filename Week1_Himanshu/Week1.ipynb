# Failure Analysis


```python
# List all failure columns
failure_types = ['TWF', 'HDF', 'PWF', 'OSF', 'RNF']

# Count each failure type
for failure in failure_types:
    count = df[failure].sum()
    percentage = (count / len(df)) * 100
    print(f"{failure}: {count} cases ({percentage:.2f}%)")
```

```python
# Bar chart of failure types
failure_counts = [df[f].sum() for f in failure_types]

plt.figure()
plt.bar(failure_types, failure_counts)
plt.title('Types of Machine Failures')
plt.xlabel('Failure Type')
plt.ylabel('Number of Cases')
plt.show()
```

# Normal and Failed Machines Count

```python
# Compare statistics of failed vs normal machines

# Separate failed and normal machines
normal_machines = df[df['Machine failure'] == 0]
failed_machines = df[df['Machine failure'] == 1]

print("Total Normal Machines:", len(normal_machines))
print("Total Failed Machines:", len(failed_machines))
```

# Statistics Report of Machines

# FAILED VS NORMAL

```python
# Compare each feature
columns = ['Air temperature [K]', 
           'Process temperature [K]',
           'Rotational speed [rpm]',
           'Torque [Nm]',
           'Tool wear [min]']



for col in columns:
    normal_avg = normal_machines[col].mean()
    failed_avg = failed_machines[col].mean()
    
    print(f"\n{col}:")
    print(f"  Normal machines average : {round(normal_avg, 2)}")
    print(f"  Failed machines average : {round(failed_avg, 2)}")
```

```python
# Visual comparison graph
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Temperature
axes[0,0].boxplot([normal_machines['Air temperature [K]'],
                   failed_machines['Air temperature [K]']],
                   labels=['Normal', 'Failed'])
axes[0,0].set_title('Air Temperature')
axes[0,0].set_ylabel('Temperature (K)')

# Torque
axes[0,1].boxplot([normal_machines['Torque [Nm]'],
                   failed_machines['Torque [Nm]']],
                   labels=['Normal', 'Failed'])
axes[0,1].set_title('Torque')
axes[0,1].set_ylabel('Torque (Nm)')

# RPM
axes[1,0].boxplot([normal_machines['Rotational speed [rpm]'],
                   failed_machines['Rotational speed [rpm]']],
                   labels=['Normal', 'Failed'])
axes[1,0].set_title('Rotational Speed')
axes[1,0].set_ylabel('RPM')

# Tool Wear
axes[1,1].boxplot([normal_machines['Tool wear [min]'],
                   failed_machines['Tool wear [min]']],
                   labels=['Normal', 'Failed'])
axes[1,1].set_title('Tool Wear')
axes[1,1].set_ylabel('Minutes')

plt.suptitle('Normal vs Failed Machines Comparison', fontsize=14)
plt.tight_layout()
plt.show()
```
