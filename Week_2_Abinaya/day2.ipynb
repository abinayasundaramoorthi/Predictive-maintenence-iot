```python
import pandas as pd
import numpy as np
df = pd.read_csv("predictive_maintainence_timestamp.csv",parse_dates=["timestamp"])
```

```python
df.head()
```

```python
df.shape
```

```python
np.random.seed(42)
```

```python
c_df=pd.DataFrame()
```

```python
c_df['timestamp'] = df['timestamp']
```

```python
c_df["shift"] = pd.cut(df["timestamp"].dt.hour,bins=[-1,7,15,23],labels=["Night","Morning","Evening"])
```

```python
c_df["production_load"]=np.random.uniform(0.6,1.0,len(df))
```

```python
c_df["power_quality_index"]=np.random.normal(0.85,1.00,len(df))
```

```python
c_df["maintenance_window"] = (df["Tool wear [min]"] // 120)
```

```python
np.random.seed(42)
```

```python
c_df["operator_experience"] = np.random.choice([1,2,3,4,5],size=len(df))
```

```python
c_df.head()
```

```python
merged_df = pd.merge(df, c_df, on='timestamp')
```

```python
merged_df.head()
```

```python
merged_df.shape
```

```python
print(merged_df.columns.isnull().sum())
```

```python
merged_df.to_csv("Fusion_dataset.csv",index=False)
```
