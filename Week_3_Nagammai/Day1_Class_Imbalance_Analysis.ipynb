```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('default')
```

```python
df = pd.read_csv("C:\\Users\\delta\\Predictive-maintenence-iot\\dataset\\ai4i2020.csv")
print("Dataset Loaded Successfully")
print("Dataset Shape:", df.shape)
df.head()
```

```python
print("Dataset Information")
print("="*50)
print("Number of Rows :", df.shape[0])
print("Number of Columns :", df.shape[1])
print("\nColumn Names")
print(df.columns.tolist())
```

```python
target = "Machine failure"
failure_counts = df[target].value_counts()
print("Target Variable Distribution")
print("="*50)
print(failure_counts)
```

```python
failure_percentage = (df[target].value_counts(normalize=True)* 100)
print("Class Percentage Distribution")
print("="*50)
print(failure_percentage.round(2))
```

```python
majority_class = failure_counts.max()
minority_class = failure_counts.min()
imbalance_ratio = majority_class / minority_class
print("="*50)
print("CLASS IMBALANCE ANALYSIS")
print("="*50)
print(f"Majority Class Count : {majority_class}")
print(f"Minority Class Count : {minority_class}")
print(f"Imbalance Ratio      : {imbalance_ratio:.2f}:1")
if imbalance_ratio > 20:
    print("\nSeverity Level: HIGHLY IMBALANCED")
elif imbalance_ratio > 10:
    print("\nSeverity Level: MODERATELY IMBALANCED")
else:
    print("\nSeverity Level: ACCEPTABLE")
print("="*50)
```

```python
plt.figure(figsize=(8,5))
ax = sns.countplot(data=df,x="Machine failure",hue="Machine failure",palette="Set2",legend=False)
plt.title("Machine Failure Class Distribution",fontsize=14,fontweight="bold")
plt.xlabel("Machine Failure")
plt.ylabel("Count")
for p in ax.patches:
    ax.annotate(f"{int(p.get_height())}",(p.get_x()+0.25, p.get_height()+50))
plt.tight_layout()
plt.show()
plt.savefig("Machine_Failure_Class_Distribution.png")
```

```python
plt.figure(figsize=(6,6))
plt.pie( failure_percentage,labels=["No Failure","Failure"],autopct="%1.2f%%",startangle=90)
plt.title("Machine Failure Percentage Distribution")
plt.show()
plt.savefig("Machine_Failure_Percentage_Distribution.png")
```

```python
summary = pd.DataFrame({"Class":[ "No Failure (0)","Failure (1)"],"Count":failure_counts.values,
    "Percentage":failure_percentage.values.round(2)
})
summary
```

```python
print("Minority Class Analysis")
print("="*50)
if imbalance_ratio > 10:
    print("Severe Class Imbalance Detected")
    print("Model may struggle to learn failure patterns.")
else:
    print("Dataset is reasonably balanced.")
```

```python
total_samples = len(df)
baseline_accuracy = (majority_class / total_samples) * 100
print("Baseline Accuracy Insight")
print("="*50)
print(f"If model predicts ONLY 'No Failure', Accuracy = {baseline_accuracy:.2f}%")
print("\nThis shows why Accuracy alone is not enough.")
```

```python
weight_failure = total_samples / (2 * minority_class)
weight_normal = total_samples / (2 * majority_class)
print("Suggested Class Weights")
print("="*50)
print(f"Normal Class Weight : {weight_normal:.3f}")
print(f"Failure Class Weight: {weight_failure:.3f}")
```
