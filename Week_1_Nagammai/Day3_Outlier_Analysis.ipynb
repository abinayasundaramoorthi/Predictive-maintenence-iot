```python
def find_outliers(column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower) | (df[column] > upper)]
    return len(outliers)
```

```python
print("Temperature Outliers:",
      find_outliers('Air temperature [K]'))
print("Torque Outliers:",
      find_outliers('Torque [Nm]'))
print("RPM Outliers:",
      find_outliers('Rotational speed [rpm]'))
```

```python
outlier_summary = pd.DataFrame({
    'Feature': [
        'Air Temperature',
        'Torque',
        'Rotational Speed'
    ],
    'Outlier Count': [
        find_outliers('Air temperature [K]'),
        find_outliers('Torque [Nm]'),
        find_outliers('Rotational speed [rpm]')
    ]
})
outlier_summary
```
