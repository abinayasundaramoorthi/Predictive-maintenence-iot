```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Infoct_project 1 sample practice\predictive_maintenance.csv')
print("Dataset loaded!")
print(df.shape)
```

```python
sns.boxplot(data=df,x="Air temperature [K]")
plt.title("Air temperature outlier detection")
plt.show()
```

```python
sns.boxplot(data=df,x="Process temperature [K]")
plt.title("Process temperature outlier detection")
plt.show()
```

```python
sns.boxplot(data=df,x="Tool wear [min]")
plt.title("Tool wear outlier")
plt.show()
```

```python
sns.boxplot(data=df,x="Rotational speed [rpm]")
plt.title("Rotational speed outlier")
plt.show()
```

```python
sns.boxplot(data=df,x="Torque [Nm]")
plt.title("Torque outlier")
plt.show()
```

```python
df["Type_encoded"]=df["Type"].map({"L":0,"M":1,"H":2})
print(df[["Type","Type_encoded"]].head(10))
```

```python
df["heat_stress_index"]=df["Air temperature [K]"] * df["Tool wear [min]"]
print(df[["Air temperature [K]","Tool wear [min]","heat_stress_index"]].head(10))
```

```python
df["speed_torque_ratio"]=df["Rotational speed [rpm]"]/df["Torque [Nm]"]
print(df[["Rotational speed [rpm]","Torque [Nm]","speed_torque_ratio"]].head(10))
```

```python
df["wear_per_rotation"]=df["Tool wear [min]"]/df["Rotational speed [rpm]"]*1000
print(df[["Tool wear [min]","Rotational speed [rpm]","wear_per_rotation"]].head(10))
```
