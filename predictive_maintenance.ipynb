```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("Libraries loaded successfully!")
```

```python
df = pd.read_csv("predictive_maintenance.csv")
print("Dataset loaded!")
```

```python
print("Shape of dataset:", df.shape)
# This tells you how many rows and columns you have
```

```python
print("Columns:", df.columns.tolist())
```

```python
# Basic info - data types, null values
df.info()

# Statistical summary - mean, min, max etc.
df.describe()
```

```python
print("=== TEMPERATURE ANALYSIS ===")
print(df['Air temperature [K]'].describe())  # Change column name if different

# Visualization
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
sns.histplot(df['Air temperature [K]'], kde=True, color='orange')
plt.title("Air Temperature Distribution")

plt.subplot(1, 2, 2)
sns.boxplot(y=df['Air temperature [K]'], color='orange')
plt.title("Temperature Boxplot")

plt.tight_layout()
plt.show()
```
