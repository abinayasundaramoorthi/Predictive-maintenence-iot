## Day 5 — Review Results & Submit Findings

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_results=pd.read_csv("threshold_results.csv")
df_results.head()
```

```python
print("=" * 70)
print("WEEK 4 - THRESHOLD TUNING RESULTS")
print("=" * 70)

print()

print(df_results[
    [
        "threshold",
        "tp",
        "fp",
        "tn",
        "fn",
        "precision",
        "recall",
        "f1"
    ]
].to_string(index=False))
```

```python
print("Recommended Threshold : 0.80")
```

```python
plt.figure(figsize=(10,5))

plt.plot(
    df_results["threshold"],
    df_results["precision"],
    marker="o",
    label="Precision"
)

plt.plot(
    df_results["threshold"],
    df_results["recall"],
    marker="o",
    label="Recall"
)

plt.plot(
    df_results["threshold"],
    df_results["f1"],
    marker="o",
    label="F1 Score"
)

plt.axvline(
    x=0.80,
    color="red",
    linestyle="--",
    label="Recommended Threshold"
)

plt.title("Precision, Recall and F1 Score Across Thresholds")

plt.xlabel("Threshold")

plt.ylabel("Score")

plt.legend()

plt.grid(True)

plt.show()
```
