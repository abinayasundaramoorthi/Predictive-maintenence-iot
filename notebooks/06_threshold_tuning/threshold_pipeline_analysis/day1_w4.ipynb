## Threshold Tuning Framework & Confusion Matrix Setup

```python
import pandas as pd
import numpy as np
from sklearn.metrics import confusion_matrix
```

```python
results = pd.read_csv("member3_day1_w4results.csv")

print("Data loaded successfully")
print(results.head())
```

```python
y_true = results["Actual"].values
y_proba = results["Failure Probability"].values

print("Shape:", y_proba.shape)
print("Min probability:", y_proba.min())
print("Max probability:", y_proba.max())
```

```python
y_pred_05 = (y_proba >= 0.5).astype(int)
print("Matches model predictions:",
      (y_pred_05 == results["Predicted"].values).all())
```

```python
def predict_at_threshold(proba, threshold):
    return (proba >= threshold).astype(int)
```

```python
def confusion_summary(y_true, y_pred, threshold):
    cm = confusion_matrix(y_true, y_pred)
    tn, fp, fn, tp = cm.ravel()

    print(f"\n--- Threshold: {threshold} ---")
    print("TN:", tn)
    print("FP:", fp)
    print("FN:", fn)
    print("TP:", tp)

    return tn, fp, fn, tp
```

```python
y_pred = predict_at_threshold(y_proba, 0.5)
confusion_summary(y_true, y_pred, 0.5)
```

```python
print(confusion_matrix(y_true, results["Predicted"]))
```
