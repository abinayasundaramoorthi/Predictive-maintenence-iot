```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_validate, StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from imblearn.pipeline import Pipeline
from imblearn.over_sampling import SMOTE
```

```python
df = pd.read_csv("C:\\Users\\delta\\Predictive-maintenence-iot\\dataset\\ai4i2020.csv")
print("Dataset Shape:", df.shape)
df.head()
```

```python
print("="*60)
print("DATASET SUMMARY")
print("="*60)
print("Rows :", df.shape[0])
print("Columns :", df.shape[1])
print("\nMissing Values")
print(df.isnull().sum())
print("\nMachine Failure Distribution")
print(df["Machine failure"].value_counts())
```

```python
target = "Machine failure"
X = df.drop(columns=["UDI","Product ID",target])
y = df[target]
X = pd.get_dummies(X,columns=["Type"],drop_first=True)
print("Prepared Feature Shape:", X.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.20,random_state=42,stratify=y)
print("Training Samples :", len(X_train))
print("Testing Samples  :", len(X_test))
```

```python
original_distribution = y_train.value_counts()
print(original_distribution)
plt.figure(figsize=(6,4))
sns.countplot(x=y_train)
plt.title("Original Training Distribution")
plt.show()
```

```python
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train,y_train)
print("SMOTE Completed Successfully")
```

```python
balanced_distribution = y_train_smote.value_counts()
print(balanced_distribution)
plt.figure(figsize=(6,4))
sns.countplot(x=y_train_smote)
plt.title("Balanced Training Distribution")
plt.show()
```

```python
comparison = pd.DataFrame({
    "Original": original_distribution,
    "SMOTE": balanced_distribution
})
comparison
```

```python
before_ratio = (original_distribution[0]/original_distribution[1])
after_ratio = (balanced_distribution[0]/balanced_distribution[1])
print("Before SMOTE Ratio :", round(before_ratio,2))
print("After SMOTE Ratio :", round(after_ratio,2))
```

```python
pipeline = Pipeline([
    ("smote", SMOTE(random_state=42)),
    ("model", LogisticRegression(max_iter=1000))
])
print("Pipeline Ready")
```

```python
cv = StratifiedKFold(n_splits=5,shuffle=True,random_state=42)
scores = cross_validate(pipeline,X_train,y_train,scoring=["accuracy","precision","recall","f1"],cv=cv)
results = pd.DataFrame({
    "Accuracy":scores["test_accuracy"],
    "Precision":scores["test_precision"],
    "Recall":scores["test_recall"],
    "F1 Score":scores["test_f1"]
})
results
```

```python
summary = pd.DataFrame({
    "Metric":[
        "Accuracy",
        "Precision",
        "Recall",
        "F1 Score"
    ],
    "Average":[
        results["Accuracy"].mean(),
        results["Precision"].mean(),
        results["Recall"].mean(),
        results["F1 Score"].mean()
    ]
})
summary
```

```python
plt.figure(figsize=(8,5))
plt.bar(summary["Metric"],summary["Average"])
plt.ylim(0,1)
plt.title("Average Cross Validation Metrics")
plt.savefig("Average_Cross_Validation_Metrics.png")
plt.show()
```
