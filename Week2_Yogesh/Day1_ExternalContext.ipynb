```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pdf = pd.read_csv(r'C:\Infoct_project 1 sample practice\dataset\predictive_maintenance.csv')
print("Dataset loaded!")
print(df.shape)
```

```python
np.random.seed(42)
df["Ambient_temperature"]=np.random.normal(loc=295,scale=3,size=len(df))
df["Load_density"]=np.random.uniform(low=0.3,high=1.0,size=len(df))
print(df[["Air temperature [K]","Ambient_temperature","Load_density"]])
```

```python
df["temperature_gap"]=df["Air temperature [K]"]-df["Ambient_temperature"]
print(df[["Air temperature [K]","Ambient_temperature","temperature_gap"]].head())
```

```python
df["Power"] = df["Rotational speed [rpm]"] * df["Torque [Nm]"]
```

```python
df["Load_stress"]=df["Power"]*df["Load_density"]
print(df[["Power","Load_density","Load_stress"]].head())
```
