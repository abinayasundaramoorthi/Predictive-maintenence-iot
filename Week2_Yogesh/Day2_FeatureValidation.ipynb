```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Infoct_project 1 sample practice\predictive_maintenance.csv')
print(df.shape)
```

```python
df["temp_difference"] = df["Process temperature [K]"] - df["Air temperature [K]"]
df["power"] = df["Rotational speed [rpm]"] * df["Torque [Nm]"]
df["tool_wear_rate"] = df["Tool wear [min]"] / df["Rotational speed [rpm]"]
print(df[["temp_difference", "power", "tool_wear_rate"]].head())
```

```python
df["heat_stress_index"] = df["Air temperature [K]"] * df["Tool wear [min]"]
df["speed_torque_ratio"] = df["Rotational speed [rpm]"] / df["Torque [Nm]"]
df["wear_per_rotation"] = df["Tool wear [min]"] / df["Rotational speed [rpm]"] * 1000
print(df[["heat_stress_index", "speed_torque_ratio", "wear_per_rotation"]].head())
```

```python
np.random.seed(42)
df['Ambient_temperature'] = np.random.normal(loc=295, scale=3, size=len(df))
df['Load_density'] = np.random.uniform(low=0.3, high=1.0, size=len(df))
df["temperature_gap"] = df["Air temperature [K]"] - df["Ambient_temperature"]
df["load_stress"] = df["power"] * df["Load_density"]
print(df[["Ambient_temperature", "Load_density", "temperature_gap", "load_stress"]].head())
```

```python
new_features = ['temp_difference', 'power', 'tool_wear_rate', 'heat_stress_index', 'speed_torque_ratio', 'wear_per_rotation','temperature_gap', 'load_stress']
for feature in new_features:
    correlation = df[feature].corr(df['Target'])
    print(f"{feature}: {correlation:.3f}")
```

```python
df["effective_power"] = df["power"] / (df["Ambient_temperature"] - 273)
print(df[["power", "Ambient_temperature", "effective_power"]].head())
```

```python
correlation = df["effective_power"].corr(df['Target'])
print(f"effective_power: {correlation:.3f}")
```
