```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score
)
print("Libraries imported successfully.")
```

```python
df = pd.read_csv("C:\\Users\\delta\\Predictive-maintenence-iot\\dataset\\ai4i2020.csv")
print("Dataset Loaded Successfully")
print("Shape:", df.shape)
df.head()
```

```python
print("Dataset Information")
print("-" * 50)
df.info()
print("\nMissing Values")
print(df.isnull().sum())
```

```python
target = "Machine failure"
print("Target Variable:", target)
print("\nClass Distribution")
print(df[target].value_counts())
```

```python
# Sensor features used in baseline model
sensor_features = [
    'Air temperature [K]',
    'Process temperature [K]',
    'Rotational speed [rpm]',
    'Torque [Nm]',
    'Tool wear [min]'
]
print("Baseline Features")
print(sensor_features)
```

```python
# Baseline dataset (sensor data only)
X_baseline = df[sensor_features]
y = df[target]
print("Baseline Dataset Shape")
print(X_baseline.shape)
X_baseline.head()
```

```python
# Train-test split for baseline experiment
Xb_train, Xb_test, yb_train, yb_test = train_test_split(
    X_baseline,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)
print("Training Shape:", Xb_train.shape)
print("Testing Shape:", Xb_test.shape)
```

```python
# Add contextual information
context_features = [
    'Air temperature [K]',
    'Process temperature [K]',
    'Rotational speed [rpm]',
    'Torque [Nm]',
    'Tool wear [min]',
    'Type'
]
print("Context Features")
print(context_features)
```

```python
# Context-aware dataset
X_context = df[context_features]
X_context.head()
```

```python
# Convert machine type into numerical format
X_context = pd.get_dummies(
    X_context,
    columns=['Type'],
    drop_first=True
)
print("Context Dataset Shape")
print(X_context.shape)
X_context.head()
```

```python
# Train-test split for context experiment
Xc_train, Xc_test, yc_train, yc_test = train_test_split(
    X_context,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)
print("Training Shape:", Xc_train.shape)
print("Testing Shape:", Xc_test.shape)
```

```python
# Experiment setup
experiments = {
    "Experiment A": "Sensor Features Only",
    "Experiment B": "Sensor + Context Features"
}

for exp, description in experiments.items():
    print(f"{exp} : {description}")
```

```python
# Metrics selected for model comparison
evaluation_metrics = [
    "Accuracy",
    "Precision",
    "Recall",
    "F1 Score",
    "ROC-AUC"
]
print("Evaluation Metrics")
for metric in evaluation_metrics:
    print("-", metric)
```
