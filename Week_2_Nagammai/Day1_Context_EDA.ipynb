```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use("ggplot")
```

```python
df = pd.read_csv("C:\\Users\\delta\\Predictive-maintenence-iot\\dataset\\ai4i2020.csv")
df.head(10)
df.info()
```

```python
df["Type"].value_counts()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="Type")
plt.title("Machine Type Distribution")
plt.savefig("machine_type_distribution.png")
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.histplot(df["Air temperature [K]"], bins=30,kde=True)
plt.title("Air Temperature Distribution")
plt.savefig("air_temperature_distribution.png")
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.histplot(df["Process temperature [K]"],bins=30,kde=True)
plt.title("Process Temperature Distribution")
plt.savefig("process_temperature_distribution.png")
plt.show()
```

```python
print("No time-based column available.")
print("Seasonality analysis cannot be performed on AI4I dataset.")
```

```python
pd.crosstab(df["Type"],df["Machine failure"])
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df,x="Type",hue="Machine failure")
plt.title("Machine Failure by Type")
plt.savefig("failure_by_type.png")
plt.show()
```
