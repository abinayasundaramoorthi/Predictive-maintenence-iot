```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Infoct_project 1 sample practice\dataset\predictive_maintenance.csv')
print("Dataset loaded!")
```

```python
df = df.dropna()
print(df.shape)
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True),annot=True,fmt=".2f",cmap="coolwarm")
plt.title("correlation heatmap - sensor features")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x="Target", data=df)
plt.title("Class Distribution - Machine Failure")
plt.xlabel("Target (0 = No Failure, 1 = Failure)")
plt.ylabel("Count")
plt.show()
```

```python
df["temp_difference"] = df["Process temperature [K]"] - df["Air temperature [K]"]
print(df[["Air temperature [K]", "Process temperature [K]", "temp_difference"]].head())
```

```python
df["tool_wear_rate"] = df["Tool wear [min]"] / df["Rotational speed [rpm]"]
print(df[["Tool wear [min]", "Rotational speed [rpm]", "tool_wear_rate"]].head())
```

```python
df["power"]=df["Rotational speed [rpm]"]*df["Torque [Nm]"]
print(df["power"].head())
```
